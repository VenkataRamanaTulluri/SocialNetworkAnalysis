```
import networkx as nx
G_fb = nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
print(nx.info(G_fb))
nx.draw_networkx(G_fb)
```

```
pos = nx.spring_layout(G_fb)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
eigCent = nx.eigenvector_centrality(G_fb)
degCent = nx.degree_centrality(G_fb)
```

```
node_size =  [v * 10000 for v in degCent.values()]
'''
Node color varies with Degree and node size with Degree Centrality
'''
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
```

```
node_size =  [v * 10000 for v in eigCent.values()]
'''
Node color varies with Degree and node size with Eigenvector Centrality
'''
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
```

```
node_size =  [v * 10000 for v in betCent.values()]
'''
Node color varies with Degree and node size with Betweenness Centrality
'''
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
```

```
print("Top 5 nodes w.r.t.:")
print("Degree Centrality -",sorted(degCent, key=degCent.get, reverse=True)[:5])
print("Eigenvector Centrality -",sorted(eigCent, key=eigCent.get, reverse=True)[:5])
print("Betweenness Centrality -",sorted(betCent, key=betCent.get, reverse=True)[:5])

print("Average Clustering:", nx.average_clustering(G_fb))
print("Eccentricity:", nx.eccentricity(G_fb))
```

```
print("Degree of node 1912:", nx.degree(G_fb, 1912))
print("Local Clustering Coefficient of node 1912:", nx.clustering(G_fb, 1912))
print("Shortest Path between nodes 1912,1684:", nx.shortest_path(G_fb, 1912, 1684))
print("Shortest Path Length between nodes 1912,1684:", nx.shortest_path_length(G_fb, 1912, 1684))
```

```
print("Subgraph:")
G_fb_subgraph = G_fb.subgraph([107, 1684, 1912, 3437, 0, 2266, 2206, 2233, 2464, 1085, 50, 110])
nx.draw_networkx(G_fb_subgraph, pos = nx.spring_layout(G_fb_subgraph))
```

```
print("Breadth First Search with source node 107:")
T = nx.bfs_tree(G_fb_subgraph, 107)
nx.draw_networkx(T, pos = nx.spring_layout(T)) 
```

```
print("Depth First Search with source node 107:")
T = nx.dfs_tree(G_fb_subgraph, 107)
print(list(nx.dfs_preorder_nodes(T, source=107)))
nx.draw_networkx(T, pos = nx.spring_layout(T))
```
